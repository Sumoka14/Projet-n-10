#### SÃ©paration du dataframe

SÃ©paration du dataframe en 2, une moitiÃ© servant Ã  entrainer le modÃ¨le, l'autre Ã  vÃ©rifier son efficacitÃ©.

```python
from sklearn.model_selection import train_test_split

df_train = df.dropna(how='any')
df_train['is_genuine'] = df_train['is_genuine'].astype('category')

feat_miss = ["diagonal", "height_left", "height_right", "margin_up", "length"]


X = df_train[feat_miss]
Y = df_train["margin_low"]


# divise l'ensemble des donnÃ©es en un set d'entraÃ®nement et un set de test avec une rÃ©partition 70% / 30%.
# Attribuez Ã  random_state une valeur quelconque, ce qui assure la cohÃ©rence.
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.32, random_state=657)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

#### Choix des colonnes Ã  garder

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


lm = LinearRegression()
lm.fit(X, Y)

rfe = RFE(lm, n_features_to_select=4)             # running RFE
rfe = rfe.fit(X, Y)

list(zip(X_train, rfe.support_, rfe.ranking_))
```

#### Entrainement du modÃ¨le

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
linear_model = LinearRegression()

linear_model.fit(X_train, Y_train)
```

#### Test du modÃ¨le

```python
import statsmodels.formula.api as sm

# sm.OLS(Y, X).fit().summary()

# linear_model.summary()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Ã©valuation du modÃ¨le pour l'ensemble d'entraÃ®nement

y_train_predict = linear_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

mae = mean_absolute_error(Y_train, y_train_predict)
print("La performance du ModÃ¨le pour le set de Training")
print("------------------------------------------------")
print("l'erreur RMSE esst {}".format(rmse))
print('le score R2 est {}'.format(r2))
print("l'erreur moyenne est {}".format(mae))
print("\n")

# Ã©valuation du modÃ¨le pour le set de tesst

y_test_predict = linear_model.predict(X_test)
# racine carrÃ©e de l'erreur quadratique moyenne du modÃ¨le
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

# score R carrÃ© du modÃ¨le
r2 = r2_score(Y_test, y_test_predict)

mae = mean_absolute_error(Y_test, y_test_predict)
print("La performance du ModÃ¨le pour le set de Test")
print("--------------------------------------------")
print("l'erreur RMSE est {}".format(rmse))
print('le score R2 score est {}'.format(r2))
print("l'erreur moyenne est {}".format(mae))
```

## ACP

```python
def correlation_graph(pca,
                      x_y,
                      features):

    # Extrait x et y
    x, y = x_y

    # Taille de l'image (en inches)
    fig, ax = plt.subplots(figsize=(10, 9))

    # Pour chaque composante :
    for i in range(0, pca.components_.shape[1]):

        # Les flÃ¨ches
        ax.arrow(0, 0,
                 pca.components_[x, i],
                 pca.components_[y, i],
                 head_width=0.07,
                 head_length=0.07,
                 width=0.02, )

        # Les labels
        plt.text(pca.components_[x, i] + 0.05,
                 pca.components_[y, i] + 0.05,
                 features[i])

    # Affichage des lignes horizontales et verticales
    plt.plot([-1, 1], [0, 0], color='grey', ls='--')
    plt.plot([0, 0], [-1, 1], color='grey', ls='--')

    # Nom des axes, avec le pourcentage d'inertie expliquÃ©
    plt.xlabel('F{} ({}%)'.format(
        x+1, round(100*pca.explained_variance_ratio_[x], 1)))
    plt.ylabel('F{} ({}%)'.format(
        y+1, round(100*pca.explained_variance_ratio_[y], 1)))

    plt.title("Cercle des corrÃ©lations (F{} et F{})".format(x+1, y+1))

    # Le cercle
    an = np.linspace(0, 2 * np.pi, 100)
    plt.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale

    # Axes et display
    plt.axis('equal')
    plt.show(block=False)
```

```python
from sklearn.preprocessing import StandardScaler

centRed = StandardScaler()

dfToScale = df.drop(columns='is_genuine')

centRed.fit(dfToScale)

dfCentreReduit = centRed.transform(dfToScale)
dfCentreReduit = pd.DataFrame(dfCentreReduit, columns=dfToScale.columns)

dfCentreReduit.describe()
```

```python
from sklearn.decomposition import PCA

features = dfCentreReduit.columns
names = dfCentreReduit.index

X = dfCentreReduit.values

n_components = 6
pca = PCA(n_components=n_components)
pca.fit(X)

scree = (pca.explained_variance_ratio_*100).round(2)
scree_cum = scree.cumsum().round()
x_list = range(1, n_components+1)

plt.bar(x_list, scree)
plt.plot(x_list, scree_cum, c="red", marker='o')
plt.xlabel("rang de l'axe d'inertie")
plt.ylabel("pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)

x_y = (0, 1)
correlation_graph(pca, x_y, features)
```

```python
from sklearn.cluster import KMeans

n_clusters = 2
n_components = 6

kmeans = KMeans(n_clusters=n_clusters, n_init=50)
kmeans.fit(dfCentreReduit)
labels = kmeans.labels_

pca = PCA(n_components=n_components)
pca.fit(dfCentreReduit)

X_proj = pca.transform(dfCentreReduit)
X_proj = pd.DataFrame(X_proj, columns=["PC"+str(i+1)
                      for i in range(n_components)])


fig, ax = plt.subplots(1, 1, figsize=(8, 7))
ax.scatter(X_proj.iloc[:, 0], X_proj.iloc[:, 1], c=labels, cmap="Set1")
ax.set_xlabel("F1")
ax.set_ylabel("F2")
plt.show()
```

## CrÃ©ation de la fonction

```python
ClusterVrai = 0
ClusterFaux = 1
```

On garde les centroÃ¯des des 2 clusters. Pour dÃ©terminer si un billet est faux, on regarde de quel centroÃ¯de il est le plus proche.

```python
X_proj['Cluster'] = kmeans.labels_
X_proj
```

```python
centroids = X_proj.groupby('Cluster').mean()
centroids
```

```python
import math


def VraiFaux_Kmeans(df, pca=pca, centroids=centroids, ClusterVrai=ClusterVrai, ClusterFaux=ClusterFaux, features=features):

    df_Etud = df[features]

    dfCentreReduit = centRed.transform(df_Etud)
    dfCentreReduit = pd.DataFrame(dfCentreReduit, columns=df_Etud.columns)

    X_proj = pca.transform(dfCentreReduit)
    X_proj = pd.DataFrame(X_proj, columns=["PC"+str(i+1)
                                           for i in range(n_components)])

    for i in range(0, len(df)):
        billet = X_proj.iloc[i]

        distVrai = math.dist(centroids.iloc[ClusterVrai], billet)
        distFaux = math.dist(centroids.iloc[ClusterFaux], billet)

        if distVrai < distFaux:
            df.at[i, 'PrÃ©diction_Kmean'] = True
        else:
            df.at[i, 'PrÃ©diction_Kmean'] = False

    return df
```
